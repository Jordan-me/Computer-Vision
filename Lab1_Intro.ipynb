## Loading an  Image

demo_RGB_image.png


<br>in order to load an image using **opencv** the method imread is used with the relative path name of the image to be loaded.  


```python
root_dir = "lab_1"
demo_image_path = os.path.join(root_dir, 'demo_RGB_image.png')
cv2_demo_image = cv2.imread(demo_image_path)
```

 Let's see some info about the image:<br> 
 * **Image type:** <code>type()</code> - as you can see in the output of the cell below, Opencv uses a numpy array to represent the image.
 *  **Dimensions:** <code>shape</code> - the image dimensions are it's  height by it's width by it's amount of channels. 
 * **Size attribute:** <code>size</code> - holds the product of image width by image height by the amount of channels 
 * **dtype**: <code>dtype</code> - type of array.

```python
print('type:{} dimensions: {}[Height, Width, Channel] size:{} [pixels], dtype:{}'
.format(type(cv2_demo_image), cv2_demo_image.shape, 
        cv2_demo_image.size, cv2_demo_image.dtype))
```

## Color Image Vs Grayscale Image: 

in this section you will see how to convert a color image to a grayscale image and display them with titles side by side 

```python
demo_image_gray = cv2.cvtColor(cv2_demo_image, cv2.COLOR_BGR2GRAY)
plt.subplot(131)
plt.imshow(demo_image_rgb)
plt.title('input color image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(demo_image_gray)
plt.title('default view image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(demo_image_gray,cmap="gray")
plt.title(' indicate gray cmap')
plt.axis('off')
```

# Load Images with PIL 

different imgage processing libraries might represent images diffrently from each other.
In the cell beolw we will compare between PIL and open-cv. 

While open-cv uses np array to represent an image, PIL has it's own way of representing images. 

```python
pil_demo_imge_png = Image.open(demo_image_path)
pil_demo_image_jpg = Image.open(os.path.join(root_dir, 'messi.jpg'))
messi_np = cv2.cvtColor(cv2.imread(os.path.join(root_dir, 'messi.jpg')), cv2.COLOR_BGR2RGB)
```

* **Image type:** <code>type()</code> - as you can see in the output of the cell below, PIL uses it's own different types to represent images,and it's dependent on the format in which the image has been saved to the memory with. 
*  **Dimensions:** <code>size</code> - the image dimensions are it's  height by it's width. 
* **mode:** <code>mode</code> - the color representation of the image    

```python
print('type: {} dimensions: {}[H, W] mode:{}'.format(type(pil_demo_imge_png), pil_demo_imge_png.size, pil_demo_imge_png.mode ))

print('type: {} dimensions: {}[H, W] mode:{}'.format(type(pil_demo_image_jpg), pil_demo_image_jpg.size, pil_demo_image_jpg.mode ))
```

As you can see PIL keeps track of the way the format in which is the image has been saved with in the machine memory
